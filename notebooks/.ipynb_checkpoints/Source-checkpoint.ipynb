```python
from pyshake.shakelib.source import Source
from pyshake.shakelib.timeutils import ShakeDateTime
import numpy as np
import matplotlib.pyplot as plt
from openquake.hazardlib.geo import Mesh
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
event = {'lat':28.230,'lon':84.731,'depth':8.2,'locstring':'NEPAL',
             'mag':7.8,'time':ShakeDateTime.utcnow()}

gridwidth = 6.0
gridheight = 5.0
xmin = event['lon'] - (gridwidth/2.0)
xmax = event['lon'] + (gridwidth/2.0)
ymin = event['lat'] - (gridheight/2.0)
ymax = event['lat'] + (gridheight/2.0)
xdim = 0.0083
ydim = 0.0083

fault = np.array([[28.427,84.614,20],
                  [27.986,86.179,20],
                  [27.469,85.880,13],
                  [27.956,84.461,13],
                  [28.427,84.614,20]])
lons = fault[:,1]
lats = fault[:,0]
depths = fault[:,2]

reference = 'Schmidt, J.J.J 1999 Journal of GeoTectonical Science'
source = Source(event,lon=lons,lat=lats,depth=depths,reference=reference)
x,y = np.meshgrid(np.arange(xmin,xmax,xdim),np.arange(ymin,ymax,ydim))
shakemap = Mesh(x,y)
dist = source.calcDistance(shakemap,method='rjb')
faultplane = source.getSurface()
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
for surface in faultplane.surfaces:
    #these all return Point objects
    topleft = surface.top_left
    topright = surface.top_right
    bottomright = surface.bottom_right
    bottomleft = surface.bottom_left
    x = np.array([topleft.longitude,topright.longitude,
         bottomright.longitude,bottomleft.longitude,topleft.longitude])
    y = np.array([topleft.latitude,topright.latitude,
        bottomright.latitude,bottomleft.latitude,topleft.latitude])
    z = np.array([topleft.depth,topright.depth,
        bottomright.depth,bottomleft.depth,topleft.depth])
    ax.plot(x,y,-z)
    plt.hold(True)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.gca()
im = plt.imshow(dist)
plt.title('Distance grid for event %s' % event['locstring']);
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(im, cax=cax);
```
